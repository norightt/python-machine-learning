Для начала, нужно создать лист с названиями всех файлов из zip с помощью .namelist().

```python
import zipfile
zip_file = zipfile.ZipFile('/content/Data.zip', 'r')
file_names = zip_file.namelist()
```

Затем создаем два датафрейма из файлов с info и marks.

```python
info_list = []

for filik in file_names[1:11]:
  df1 = pd.read_csv(filik)
  info_list.append(df1)

df_info = pd.concat(info_list, axis = 0)

```

```python
marks_list = []

for filik in file_names[11:]:
  df2 = pd.read_csv(filik)
  marks_list.append(df2)

df_marks = pd.concat(marks_list, axis = 0)
```

И объединяем два датафрейма по колонке index, в итоге поолучается датафрейм со всей имеющийся информацией о студентах.

```python
df = pd.merge(df_info, df_marks, on="index")
```

#### Задание 16. Переведем баллы в оценки

### Система перевода баллов в оценки
####    больше 90 = A
####      80-90 = B
####      70-80 = C
####      60-70 = D
####      50-60 = E
####    меньше 50 = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец Grade и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

**В случае, если средний балл попадает на границу между оценками (т.е. равен ровно 60, 70 или 80 баллов), вы можете интерпретировать условие на своё усмотрение (т.е. можете поставить за 60 баллов оценку D, а можете - E).**

```python
def GetGrade(average_mark):
    if average_mark >= 90:
        return "A"
    elif 80 <= average_mark < 90:
        return "B"
    elif 70 <= average_mark < 80:
        return "C"
    elif 60 <= average_mark < 70:
        return "D"
    elif 50 <= average_mark < 60:
        return "E"
    else:
        return "F"

df['average_mark'] = (df['math score'] + df['reading score'] + df['writing score']) / 3

df['Grade'] = df['average_mark'].apply(GetGrade)

df = df.drop("average_mark", axis=1)
```
